```python
total_players = purchase_data["SN"].nunique()

#List of purchases by gender
Male_Purchases = purchase_data.loc[purchase_data["Gender"] == "Male"]
Female_Purchases = purchase_data.loc[purchase_data["Gender"] == "Female"]
Other_Purchases = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed"]

#Individuals by Gender
male_total = Male_Purchases['SN'].nunique()
female_total = Female_Purchases["SN"].nunique()
other_total = Other_Purchases["SN"].nunique()

#Gender Pct
male_pct = round((male_total/total_players*100), 2)
female_pct = round((female_total/total_players*100), 2)
other_pct = round((other_total/total_players*100), 2)

gender = ["Male", "Female", "Other/Non-Disclosed"]
gender_demo_df = pd.DataFrame(({
    "Total Players": [male_total, female_total, other_total],
    "Percentage of Players": [male_pct, female_pct, other_pct],
}), index=gender)

gender_demo_df
```

```python
#Total Spent by all gamers
Total_Revenue = purchase_data['Price'].sum()


#List of purchases by gender
Male_Purchases = purchase_data.loc[purchase_data["Gender"] == "Male"]
Female_Purchases = purchase_data.loc[purchase_data["Gender"] == "Female"]
Other_Purchases = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed"]

#Individuals by Gender
male_total = Male_Purchases['SN'].nunique()
female_total = Female_Purchases["SN"].nunique()
other_total = Other_Purchases["SN"].nunique()

#Summing up total spent by gender
Male_Revenue = Male_Purchases['Price'].sum()
Female_Revenue = Female_Purchases['Price'].sum()
Other_Revenue = Other_Purchases['Price'].sum()

#Number of purchases by gender
Male_Purchase_Count = len(purchase_data.loc[(purchase_data["Gender"] == 'Male')])
Female_Purchase_Count = len(purchase_data.loc[(purchase_data["Gender"] == 'Female')])
Other_Purchase_Count = len(purchase_data.loc[(purchase_data["Gender"] == 'Other / Non-Disclosed')])

#Average Purchase Price
Male_Avg_Purchase= round((Male_Revenue/Male_Purchase_Count),2)
Female_Avg_Purchase= round((Female_Revenue/Female_Purchase_Count),2)
Other_Avg_Purchase= round((Other_Revenue/Other_Purchase_Count),2)

#Average Total Purchases Per person by gender
Male_Avg_Total_Purchases= round((Male_Revenue/male_total),2)
Female_Avg_Total_Purchases=round((Female_Revenue/female_total),2)
Other_Avg_Total_Purchases=round((Other_Revenue/other_total),2)

Index = ["Male", "Female", "Other / Non-Disclosed"]

Purchase_Table = pd.DataFrame(({"Purchase Count": [Female_Purchase_Count, Male_Purchase_Count, Other_Purchase_Count], 
                             "Average Purchase Price": [Female_Avg_Purchase, Male_Avg_Purchase, Other_Avg_Purchase],
                                "Total Purchase Value": [Female_Revenue, Male_Revenue, Other_Revenue],
                                "Avg Total Purchase per Person": [Female_Avg_Total_Purchases, Male_Avg_Total_Purchases, Other_Avg_Total_Purchases]
                              }),index=Index)
Purchase_Table.head()
```

## Purchasing Analysis (Age)

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 46]
       
group_labels = ["<10","10-14", "15-19", "20-24","25-29", "30-34", "35-39", "40+"]
```

```python
pd.cut(purchase_data["Age"], bins, labels=group_labels).head()
```

```python
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_labels)
purchase_data.head()
```

```python
total_players = purchase_data["SN"].nunique()


under10_Purchases = purchase_data.loc[purchase_data["Age Group"] == "<10"]
under10_Purchases.count()
under10_Revenue = under10_Purchases['Price'].sum()
under10_Revenue
under10_players = under10_Purchases['SN'].nunique()
under10_players
ppa_under10 = round((under10_Revenue/under10_players),2)
ppa_under10
under10_player_pct = round((under10_players/total_players*100), 2)
under10_player_pct

Purchases10_14 = purchase_data.loc[purchase_data["Age Group"] == "10-14"]
Purchases10_14.count()
Revenue10_14 = Purchases10_14['Price'].sum()
Revenue10_14
players10_14 = Purchases10_14['SN'].nunique()
players10_14
ppa10_14 = round((Revenue10_14/players10_14),2)
ppa10_14
players10_14_pct = round((players10_14/total_players*100), 2)

Purchases15_19 = purchase_data.loc[purchase_data["Age Group"] == "15-19"]
Purchases15_19.count()
Revenue15_19 = Purchases15_19['Price'].sum()
Revenue15_19
players15_19 = Purchases15_19['SN'].nunique()
players15_19
ppa15_19 = round((Revenue15_19/players15_19),2)
ppa15_19
players15_19_pct = round((players15_19/total_players*100), 2)
players15_19_pct

Purchases20_24 = purchase_data.loc[purchase_data["Age Group"] == "20-24"]
Purchases20_24.count()
Revenue20_24 = Purchases20_24['Price'].sum()
players20_24 = Purchases20_24['SN'].nunique()
ppa20_24 = round((Revenue20_24/players20_24),2)
players20_24_pct = round((players20_24/total_players*100), 2)
players20_24_pct

Purchases25_29 = purchase_data.loc[purchase_data["Age Group"] == "25-29"]
Purchases25_29.count()
Revenue25_29 = Purchases25_29['Price'].sum()
players25_29 = Purchases25_29['SN'].nunique()
ppa25_29 = round((Revenue25_29/players25_29),2)
players25_29_pct = round((players25_29/total_players*100), 2)

Purchases30_34 = purchase_data.loc[purchase_data["Age Group"] == "30-34"]
Purchases30_34.count()
Revenue30_34 = Purchases30_34['Price'].sum()
players30_34 = Purchases30_34['SN'].nunique()
ppa30_34 = round((Revenue30_34/players30_34),2)
players30_34_pct = round((players30_34/total_players*100), 2)

Purchases35_39 = purchase_data.loc[purchase_data["Age Group"] == "35-39"]
Purchases35_39.count()
Revenue35_39 = Purchases35_39['Price'].sum()
players35_39 = Purchases35_39['SN'].nunique()
ppa35_39 = round((Revenue35_39/players35_39),2)
players35_39_pct = round((players35_39/total_players*100), 2)

Purchases_over40 = purchase_data.loc[purchase_data["Age Group"] == "40+"]
Purchases_over40.count()
Revenue_over40 = Purchases_over40['Price'].sum()
players_over40 = Purchases_over40['SN'].nunique()
ppa_over40 = round((Revenue_over40/players_over40),2)
players_over40_pct = round((players_over40/total_players*100), 2)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spenders = purchase_data.groupby(["SN"])
total_spend = spenders["Price"].sum()
total_spend_df = pd.DataFrame([total_spend])
total_spend_df = total_spend_df.T
total_spend_df
total_spend_renamed = total_spend_df.rename(columns={"Price": "Total Purchase Value"})
total_spend_renamed
total_transactions = spenders["Purchase ID"].count()
total_transactions_df = pd.DataFrame([total_transactions])
total_transactions_df = total_transactions_df.T
total_transactions_renamed = total_transactions_df.rename(columns={"Purchase ID": "Total Purchase Count"})
total_transactions_renamed
avg_spend = spenders["Price"].mean()
avg_spend_df = pd.DataFrame([avg_spend])
avg_spend_df = avg_spend_df.T
avg_spend_renamed = avg_spend_df.rename(columns={"Price": "Average Purchase Price"})
avg_spend_renamed

spender_frames = [total_transactions_renamed, avg_spend_renamed, total_spend_renamed]
spender_index = ["SN", "Average Purchase Price", "Total Purchase Value"]
spender_result_df = pd.concat(spender_frames, axis=1, join='inner', join_axes=None, ignore_index=False,
          keys=None, levels=None, names=None, verify_integrity=False,
          copy=True)
spender_result_df
spender_result_df = spender_result_df.sort_values('Total Purchase Value', ascending = False)
spender_result_df.head()

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_items_df = pd.DataFrame(purchase_data)
top_items_grp = top_items_df.groupby(['Item ID','Item Name'])
top_items_df = pd.DataFrame({"Purchase Count":top_items_grp["Price"].count(), 
                             "Item Price":top_items_grp["Price"].mean(),
                             "Total Purchase Value":top_items_grp["Price"].sum()})


top_items_df = top_items_df.sort_values("Purchase Count", ascending=False)
top_items_df["Item Price"] = top_items_df["Item Price"]
top_items_df["Total Purchase Value"] = top_items_df["Total Purchase Value"]

top_items_df.head()
```
